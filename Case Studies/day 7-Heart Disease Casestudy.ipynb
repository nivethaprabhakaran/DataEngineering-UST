```python
#Load and Explore the Dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/Administrator/Downloads/Heart Disease UCI.csv")
df.head(10) #displaying first 10 dataset
```

```python
df.dtypes #checking datatypes of each column
```

```python
df.isnull().sum() #checking count of null values in each column
```

```python
#summary statistics
df.describe()
```

```python
# Gender Distribution Analysis
count=df['sex'].value_counts()
m=count.get(1, 0)
f=count.get(0, 0)
print(f"Males:{m}, Females:{f}")

tot=m+f
per=np.array([m,f])/tot*100
# bar chart
labels = ['Male','Female']
plt.bar(labels, per, color=['blue','red'])
plt.ylabel('Percentage')
plt.title('Gender percentage distribution')
plt.show()
```

```python
#Age Analysis 
mina=df["age"].min()
maxa=df["age"].max()
meana=df["age"].mean()
meda=df["age"].median()
print("Minimum Age: ",mina)
print("Maximum Age: ",maxa)
print("Mean Age: ",meana)
print("Median Age: ",meda)
plt.hist(df["age"], bins=10, color="lightgray", edgecolor="black")
plt.xlabel("Age")
plt.show()
```

```python
#Target Variable Analysis 
wh=df[df["condition"]==1].value_counts().count()
woh=df[df["condition"]==0].value_counts().count()
a=[wh,woh]
plt.pie(a,labels=["With Heart Disease","Without Heart Disease"],autopct="%2.2f%%")
plt.title("Heart Disease Distribution")
plt.legend()
plt.show()
```

```python
#Correlation Between Age and Cholesterol 
cor=df[["age","chol"]].corr()
print("Correlation matrix:")
print(cor)
corv = cor.loc["age","chol"]
print(f"Correlation(age vs chol): {corv:.3f}")
# colors = ["red" if x == 1 else "blue" for x in df["condition"]]
g = df.plot.scatter(x="age", y="chol", color="blue", title="Age vs Cholesterol")
plt.xlabel("Age")
```

```python
#Chest Pain Type vs Disease 
cpd= df.groupby('cp')['condition'].agg(['sum', 'count'])
cpd['rate']=cpd['sum'] / cpd['count']
print('Disease rate by chest pain type:')
print(cpd)
# bar chart
cpd['rate'].plot(kind='bar', color='orange', figsize=(8,5))
plt.xlabel('Chest Pain Type')
plt.ylabel('Disease Rate')
plt.title('Disease Rate by Chest Pain Type')
plt.xticks(rotation=0)
plt.show()
```

```python
#Average Cholesterol by Gender
achol=df.groupby("sex")["chol"].mean()
print("Average cholesterol by gender:")
print(achol)
# bar plot
achol.plot(kind="bar", color="blue", figsize=(6, 4))
plt.xlabel("Sex f=0 m=1")
plt.ylabel("Avg Cholesterol")
plt.title("Avg Cholesterol by Gender")
plt.xticks(rotation=0)
plt.show()
```

```python
#Resting Blood Pressure Analysis
avgbp = df["trestbps"].mean()
print(f"Average BP: {avgbp:.2f}")

highbp = df[df["trestbps"] > 140]
tot = len(highbp)

dh = highbp["condition"].sum()
drh = dh / tot if tot > 0 else 0

nbp = df[df["trestbps"] <= 140]
drn = nbp["condition"].sum() / len(nbp)

print(f"Disease rate (BP > 140): {drh:.3f}")
print(f"Disease rate (BP <= 140): {drn:.3f}")

# bar chart 
rates = [drn, drh]
labels = ["BP <= 140", "BP > 140"]
plt.bar(labels, rates, color=["green", "red"])
plt.ylabel("Disease Rate")
plt.title("Resting Blood Pressure Analysis ")
plt.show()
```

```python
#Maximum Heart Rate vs Disease
disease=df[df["condition"]== 1]
nodisease=df[df["condition"]== 0]
# boxplot
plt.boxplot([nodisease["thalach"],disease["thalach"]],tick_labels=["No Disease","Disease"])
plt.ylabel("Max Heart Rate")
plt.title("Max Heart Rate vs Disease")
plt.show()
```

```python
#Exercise Induced Angina Impact
exang1=df[df["exang"]==1]
exang0=df[df["exang"]==0]

rate1=exang1["condition"].sum()/len(exang1)*100
rate0=exang0["condition"].sum()/len(exang0)*100
# bar chart
plt.bar(["exang=0","exang=1"],[rate0, rate1],color=["blue", "red"])
plt.ylabel("Disease Percentage")
plt.title("Exercise Induced Angina Impact")
plt.show()
```

```python
# ST Depression (oldpeak) Analysis 
healthy= df[df["condition"] == 0]["oldpeak"]
sick= df[df["condition"] == 1]["oldpeak"]
# histograms
fig, ax = plt.subplots(1, 2, figsize=(10,5))  # 1 row, 2 columns
# Histogram for healthy (no disease)
ax[0].hist(healthy, bins=15, alpha=0.6, label="No Disease", color="blue")
ax[0].set_title("ST Depression Analysis(oldpeak) (No Disease)")
ax[0].set_xlabel("ST Depression")
ax[0].set_ylabel("Frequency")
ax[0].legend()
# Histogram for sick (with disease)
ax[1].hist(sick, bins=15, alpha=0.6, label="Disease", color="red")
ax[1].set_title("ST Depression Analysis(oldpeak) (Disease)")
ax[1].set_xlabel("ST Depression")
ax[1].set_ylabel("Frequency")
ax[1].legend()
```

```python
# Number of Major Vessels (ca) Impact 
cad=df.groupby("ca")["condition"].mean()
print(cad)
# line chart
plt.plot(cad.index,cad.values,marker="o",color="red",linewidth=2)
plt.xlabel("Number of Major Vessels")
plt.ylabel("Disease Probability")
plt.title("Number of Major Vessels (ca) Impact ")
plt.show()
```

```python
#Thalassemia vs Disease 
crosstab=pd.crosstab(df["thal"],df["condition"])
pct=crosstab.div(crosstab.sum(axis=1), axis=0)*100
# stacked bar chart
pct.plot(kind="bar", stacked=True, color=["green", "red"])
plt.xlabel("Thalassemia Type")
plt.ylabel("Percentage")
plt.title("Thalassemia vs Disease")
plt.legend(["No Disease", "Disease"])
plt.xticks(rotation=0)
plt.show()
```

```python
#Multi-Factor Risk Analysis
highrisk=df[(df["age"]>50) & (df["chol"]>240) & (df["trestbps"]>140)]
total=len(highrisk)
diseased=highrisk["condition"].sum()
pct=diseased/total*100 if total>0 else 0
print(f"High-risk patients: {total}")
print(f"Disease percentage: {pct:.2f}")

allpct=df["condition"].sum()/len(df)*100
print(f"Overall disease percentage: {allpct:.2f}")

# bar chart
groups=["All Patients", "High-Risk Group"]
rates=[allpct, pct]
plt.bar(groups, rates, color=["blue", "red"])
plt.ylabel("Disease Percentage")
plt.title("Disease Rate: All vs High-Risk")
plt.show()
```

```python
#Create Risk Score (Custom Analysis
df["risk_score"]=(df["chol"]/200)+(df["trestbps"]/120)+df["oldpeak"]
def classify_risk(score):
    if score<2:
        return "Low"
    elif score<3:
        return "Medium"
    else:
        return "High"

df["risk_level"]=df["risk_score"].apply(classify_risk)
# distribution
risk_counts=df["risk_level"].value_counts()
print(risk_counts)

# bar chart
plt.bar(risk_counts.index, risk_counts.values, color=["green", "yellow", "red"])
plt.xlabel("Risk Level")
plt.ylabel("Count")
plt.title("Patient Risk Distribution")
plt.show()
```

1) Does cholesterol strongly impact heart disease?

Cholesterol does show a positive relationship with disease ,patients with higher levels (especially >240mg/dL) have a noticeably greater rate of heart disease. Its not the only factor, but it is a clear risk marker in the data

2) Does exercise-induced angina significantly increase risk?

Yes,when exang = 1 (chest pain during exercise) the disease percentage jumps dramatically (around 70% vs ~20%). That symptom is one of the strongest individual warning signs.


3) Is male population more vulnerable?

Yes.The male group in the dataset has a higher proportion of disease cases than females, 
so the male population appears more vulnerable given these records


4) Which feature has strongest correlation with diseasease?

Looking at the correlation matrix, the feature most tightly linked to target tended to be chest pain type or number of major vessels (ca) with high positive correlation values. In general, variables related to chest pain/vasculature showed the strongest linear association with disease.
