```python
#assessment hcf of two numbers
def hcf(a,b):
    while b:
        a,b=b,a%b
    return a
r=hcf(200,400)
print(r)
```

```python
#using function and filter to print only positive numbers 
a=[1,-3,5,-2,8,9]
def checkp(n):
    if n>0:
        return n
l=list(filter(checkp,a))
print(l)
```

```python
#using filter and lambda to print only positive numbers in a range
a=[1,-3,5,-2,8,9]
l=list(filter(lambda n:n>0,a))
print(l)
```

```python
class sports:
    def __init__(self,model,color,cost):
        self.__model=model#private variables
        self.__color=color
        self.__cost=cost
    def getmo(self):
        return self.__model
    def setmo(self,mo):
        self.__model=mo
    def getcol(self):
        return self.__color
    def setcol(self,col):
        self.__color=col

class cricket(sports):
    def __init__(self,model,color,cost,brand):
        super().__init__(model,color,cost)
        self.__brand=brand
    def getbr(self):
        return self.__brand
    def setbr(self,br):
        self.__brand=br
c=cricket("A1","Blue",7373,"Nike")
print(c.getmo()) #we can access parent class methods also
print(c.getbr())
```

```python
#simple calculator 
while(True):
    n1=int(input("Enter first number"))
    n2=int(input("Enter second number"))
    o=input("Enter the operators\nAddition = +\nSubtraction = -\nDivision = /\nMultiplication = *\ne for exit")
    if o=="+":
        r=n1+n2
        print(f"Addition of {n1} and {n2} is {r}")
    elif o=="-":
        r=n1-n2
        print(f"Subtraction of {n1} and {n2} is {r}")
    elif o=="*":
        r=n1*n2
        print(f"Multiplication of {n1} and {n2} is {r}")
    elif o=="/":
        try:
            r=n1/n2
            print(f"Division of {n1} and {n2} is {r}")
        except ZeroDivisionError as e:
            print(e)
    elif o=="e":
        print("exiting calculator..")
        break
    else:
        print("Invalid operation")
    
        
```
