```python
#set operations
s1={1,2,3,4,7,8}
s2={2,3,4,9,10}
print(s1 | s2) #s1.union(s2)
print(s1 & s2) #s1.intersection(s2)
print(s1 - s2) #s1.difference(s2)
print(s1 ^ s2) #s1.symmetric_difference(s2))
print({2,3}.issubset(s2))
```
