```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sample data representing monthly website traffic 
month=["jan","feb","mar","apr","may","jun","jul","aug","sep","oct","nov","dec"]
traffic=[150,29,99,200,120,100,60,80,97,78,122,200]
plt.plot(month,traffic)#x axis,y axis   line plot
plt.show()
```

```python
month=["jan","feb","mar","apr","may","jun","jul","aug","sep","oct","nov","dec"]
traffic=[150,29,99,200,120,100,60,80,97,78,122,200]
plt.plot(month,traffic)#x axis,y axis
#adding labels and titles
plt.xlabel("Month")
plt.ylabel("Monthly Traffic")
plt.title("Monthly Website Traffic")
plt.show()
```

```python
plt.plot(month,traffic,marker="*",markersize=18,color="blue",linestyle="-",linewidth=3)#x axis,y axis
#adding labels and titles
plt.xlabel("Month")
plt.ylabel("Monthly Traffic")
plt.title("Monthly Website Traffic")
plt.show()
```

```python
month=["jan","feb","mar","apr","may","jun","jul","aug","sep","oct","nov","dec"]
prod_a=[50,45,60,25,70,55,90,80,50,40,60,90]
prod_b=[30,90,55,80,40,60,70,50,80,40,20,40]
plt.plot(month,prod_a,marker="*",linestyle="-",color="red",label="Product A")
plt.plot(month,prod_b,marker="^",linestyle="-",label="Product B",color="blue")
#adding labels and titles
plt.xlabel("Month")
plt.ylabel("Monthly Revenue") 
plt.title("Monthly Revenue Comparison")
plt.legend()
plt.show()
```

```python
year=[2020,2021,2022,2023,2024,2025]
comp_A=[10,15,20,25,30,40]
comp_B=[10,12,15,18,20,30]
plt.plot(year,comp_A,marker="*",linestyle="-",color="red",label="Company A")
plt.plot(year,comp_B,marker="^",linestyle="--",label="Company B",color="blue")
#adding labels and titles
plt.xlabel("Year")
plt.ylabel("Salary Hike") 
plt.title("Salary Hike Comparison")
plt.legend() #used for displaying that label company A,company B
plt.show()
```

```python
categories=["category A","category B","category C","category D"]
values=[20,40,55,45]
plt.bar(categories,values,color=["red","g","blue"],width=0.5)  #color you give three color then it repeats r,g,b,r,g,b if red and blue alone its like r,b,r,b
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot")
plt.show()
```

```python
categories=["category A","category B","category C","category D"]
values=[20,40,55,45]
plt.barh(categories,values,color=["red","g","blue"],height=0.5) #barh is bar horizontal and instead of width use height
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot")
plt.show()
```
