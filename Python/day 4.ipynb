Iterator 

Iterator is an object that allows you to traverse through all the elements of a collection(like lists or tuples) one at a time.

* implements __iter__() and __next__() methods
* Raises stop iteration when no more data is found

```python
class Mynum:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        if self.num<=5:
            x=self.num
            self.num+=1
            return x
        else:
            raise StopIteration
ob=Mynum()
myiter=iter(ob)
for x in myiter:
    print(x)
```

```python
class Myeg:
    def __iter__(self):
        self.num=2
        return self
    def __next__(self):
        if self.num<=10:
            x=self.num
            self.num+=2
            return x
        else:
            raise StopIteration
ob=Myeg()
myrange=iter(ob)
for i in myrange:
    print(i)
```

Generator


A generator is a special type of iterator that is defined using a function and the yield keyword

* uses yield instead of return
* Generates values lazily(one at a time)
* more memory efficient than normal functions
  

```python
def my_generator():
    for i in range(1,6):
        yield i
gen=my_generator()
for i in gen:
    print(i)
```

Iterator and generator comparison

* Iterator : Requires class with __iter__() and __next__()
* Generator : Uses function with yield keyword
* Iterator : More code,manual implementation
* Generator : Simple and concise implementation
* Iterator : Slightly more control
* Generator : More memory efficient

```python
import time
st=time.perf_counter()
def my_generator():
    for i in range(1,11):
        yield i
gen=my_generator()
for i in gen:
    print(i,end=" ")
sto=time.perf_counter()
print(f"Execution time is {sto-st:.6f}")
```
