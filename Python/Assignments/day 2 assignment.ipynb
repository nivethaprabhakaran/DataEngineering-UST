1)why local variable cannot be called from outside the funtion

A local variable cannot be called from outside function because of variable scope and memory management.variables created inside a function are confined to that functions local scope.once the function finishes running these variables are immediately deleted from memory.which makes a program more efficient .

2)where do you use map,reduce and filter

map() used to transform every item in a list 
filter() used to extract specific items that meet the condition
reduce() used to combine all items into a single final value


```python
nums=[1,2,3]
doubled=list(map(lambda i:i*2,nums))
print(doubled)
evens=list(filter(lambda i:i%2==0,nums))
print(evens)
from functools import reduce
total=reduce(lambda i,j:i+j,nums)
print(total)
```
