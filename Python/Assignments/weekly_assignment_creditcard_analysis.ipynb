 Bank & Customer Credit Card Analysis

Analyze bank credit card features and customer spending to recommend suitable credit cards based on eligibility and potential cashback.

Import Libraries and Load Data

```python
import pandas as pd

bank_df = pd.read_excel("BankData (3).xlsx")
customer_df = pd.read_csv("customer_expense_dataset_1500.csv")

bank_df.head()
```

```python
customer_df.head()
```

Data Cleaning

Basic cleaning to ensure column names are simple and missing values are handled.

```python
#checking missing values and filling them with forward fill method
bank_df.isnull().sum()
bank_df.fillna(method="ffill",inplace=True)
bank_df.isnull().sum()
```

```python
customer_df.isnull().sum()
```

```python
#renaming column names 
bank_df.columns = [
"Bank",
"Card",
"Annual_Fee",
"Grocery_CB",
"Travel_CB",
"Dining_CB",
"Min_Income",
"Monthly_Limit",
"Grocery_Discount",
"Travel_Discount",
"Dining_Discount",
"Rewards",
"Review",
"Card_Type",
"Lounge_Access",
"Total_CB_dataset"
]

customer_df.columns = [
"Name",
"Grocery",
"Travel",
"Electricity",
"Internet",
"EMI",
"Dining",
"Entertainment",
"OTT",
"Ecommerce",
"UPI",
"Income"
]
```

Merge Bank and Customer Data

Merge datasets to evaluate credit card benefits based on customer spending patterns.

```python
df = customer_df.merge(bank_df, how="cross")
df.head()
```

```python
#renaming bank dataset column names for better understanding
bank_df.columns = [
"Bank",
"Card",
"Annual_Fee",
"Grocery_CB",
"Travel_CB",
"Dining_CB",
"Min_Income",
"Monthly_Limit",
"Grocery_Discount",
"Travel_Discount",
"Dining_Discount",
"Rewards",
"Review",
"Card_Type",
"Lounge_Access",
"Total_CB"
]
```

1 ) Best Credit Card for Grocery Shoppers

Identify credit cards offering the highest cashback for grocery purchases

```python
bank_df.sort_values("Grocery_CB", ascending=False)[
["Bank","Card","Grocery_CB"]
].head()
```

2 ) Cards with No Annual Fee

Find credit cards that do not charge any annual fee.

```python
bank_df[bank_df["Annual_Fee"] == 0][
["Bank","Card","Annual_Fee"]
]
```

3 ) Cards Suitable Based on Income

Filter cards that customers are eligible for based on their annual income.

```python
eligible = df[(df["Income"] * 12) >= df["Min_Income"]]

eligible_cards = (
    eligible.groupby("Name")["Card"]
    .apply(lambda x: ", ".join(sorted(set(x))))
    .reset_index(name="Eligible_Cards")
)

eligible_cards
```

4 ) Calculate Total Cashback

Calculate total cashback earned by customers based on their spending in grocery, travel, and dining categories.

```python
df["Total_Cashback"] = (
df["Grocery"] * df["Grocery_CB"]/100 +
df["Travel"] * df["Travel_CB"]/100 +
df["Dining"] * df["Dining_CB"]/100
)

df[["Name","Bank","Card","Total_Cashback"]]
```

5 ) Best Dining Cashback Cards

 Identify credit cards that provide the highest cashback percentage for dining expenses.

```python
bank_df.sort_values("Dining_CB", ascending=False)[
["Bank","Card","Dining_CB"]
].head()
```

6 ) Compare Annual Fee and Cashback

Compare credit card annual fees with the cashback customers can earn.

```python
fee_vs_cashback = df[["Name","Bank","Card","Annual_Fee","Total_Cashback"]]

fee_vs_cashback.sample(10)
```

7 ) Cards Suitable Based on Monthly Limit

Identify cards whose monthly credit limit can support customer spending.

```python
df["Total_Spending"] = df["Grocery"] + df["Travel"] + df["Dining"]

valid_cards = df[df["Monthly_Limit"] >= df["Total_Spending"]]

best_card = (
    valid_cards.sort_values("Monthly_Limit")
    .groupby("Name")
    .first()
    .reset_index()
)

best_card[["Name","Bank","Card","Monthly_Limit","Total_Spending"]]
```

8 ) Cards with Highest Reward Points

Identify credit cards that offer the highest reward points, helping customers maximize reward benefits.

```python
bank_df.sort_values("Rewards", ascending=False)[
["Bank","Card","Rewards"]
].head(10)
```

9 ) Cards Offering Lounge Access

Find credit cards that provide airport lounge access, which is a premium travel benefit.

```python
bank_df[bank_df["Lounge_Access"] > 0][
["Bank","Card","Lounge_Access"]
]
```

10 ) Final Credit Card Recommendation

Recommend the best credit card for each customer based on the maximum cashback they can earn from their spending.

```python
# Step 1: keep only cards customer is eligible for
eligible_cards = df[(df["Income"] * 12) >= df["Min_Income"]]

# Step 2: choose card with highest cashback for each customer
best_cards = (
    eligible_cards.sort_values("Total_Cashback", ascending=False)
    .groupby("Name")
    .first()
    .reset_index()
)

best_cards[["Name","Income","Bank","Card","Total_Cashback"]].head(40)
```
