Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
data=np.random.random_sample(5)#generate random data for histogram
print(data)
plt.hist(data,color="green",edgecolor="yellow",bins=5)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram Plot Example")
plt.show()
```

Scatter Plot

```python
x=[1,2,3,4,5]
y=[10,15,13,17,20]
plt.scatter(x,y,marker="o",color="blue")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

Piechart

```python
categories=["Housing","Transportation","Food","Entertainment","Utilities"]
expenses=[1200,300,400,200,150]
plt.figure(figsize=(5,16))
plt.pie(expenses,labels=categories,autopct="%2.2f%%")
plt.title("Monthly Expenses by Category")
plt.legend()
plt.show()
```

BoxPlot

```python
import matplotlib.pyplot as plt
data=[12,15,10,17,18,20,22,23,25,27,30,35,100,-34]
plt.boxplot(data)
plt.title("Boxplot Example")
plt.xlabel("Data")
plt.ylabel("Values")
plt.show()
```

Stack Plot

```python
quaters=["Q1","Q2","Q3","Q4"]
electronics=[10000,12000,11000,10500]
clothing=[500,6000,7500,8000]
home_appliances=[3000,7500,8200,9000]
plt.figure(figsize=(10,6))
plt.stackplot(quaters,electronics,clothing,home_appliances,labels=["Electronics","Clothings","Home Appliances"])
plt.xlabel("Quarters")
plt.ylabel("Sales")
plt.title("Product Category Sales Over Quaters")
plt.legend()
plt.show()
```

3D scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#Generate data
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2+y**2))

#3d scatter plot
fig=plt.figure()
ax=fig.add_subplot(111,projection="3d")
ax.scatter(x,y,z,c=z,cmap="viridis")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#generate data
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(-x)

#creates 2x2 subplot
fig,axs=plt.subplots(2,2,figsize=(10,8))

#plot data on each subplot
axs[0,0].plot(x,y1,"r")#top-left subplot
axs[0,0].set_title("Sine Wave")

axs[0,1].plot(x,y2,"b") #top-right subplot
axs[0,1].set_title("Cosine Wave")

axs[1,0].plot(x,y3,"g") #bottom-left subplot
axs[1,0].set_title("Tangent Wave")

axs[1,1].plot(x,y4,"m") #bottom-right subplot
axs[1,1].set_title("Exponential Wave")

#Adjust Layout
plt.tight_layout()
plt.show()
```
